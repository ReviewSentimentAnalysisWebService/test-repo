```python
from gensim.models import Word2Vec
```

```python
w2vmodel_windowsize2 = Word2Vec.load("./w2vmodel_notebook_2")
w2vmodel_windowsize3 = Word2Vec.load("./w2vmodel_notebook_3")
```

```python
# keywords = ['그립감', '기능', '무게', '클릭', '디자인', '가성비']
keywords = ["디자인", "화면", "무게", "발열", "사양", "배터리", "배송"]
```

```python
# window_size가 2인 모델의 유사단어
for k in keywords:
    print(w2vmodel_windowsize2.most_similar(positive=[k], topn=5), "\n")
```

```python
# window_size가 3인 모델의 유사단어
for k in keywords:
    print(w2vmodel_windowsize3.most_similar(positive=[k], topn=5), "\n")
```
